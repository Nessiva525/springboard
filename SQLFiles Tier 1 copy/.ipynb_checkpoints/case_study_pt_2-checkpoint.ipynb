__/* Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! */__

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        with cte AS(
            SELECT f.name AS facility_name,
            b.memid,
            b.facid AS facid,
            b.slots,
            f.membercost,
            f.guestcost,
            CASE WHEN m.memid = 0 THEN (b.slots* f.guestcost)
                when m.memid <> 0 THEN (b.slots * f.membercost) END AS revenue
        FROM Bookings b 
        LEFT JOIN Facilities f
        USING(facid)
        LEFT JOIN Members m 
        USING(memid))
    SELECT facility_name, SUM(revenue) as total_revenue
    FROM cte
    GROUP BY facid
    HAVING SUM(revenue) < 1000
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks for Q10")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT CONCAT_WS(", ", m.surname, m.firstname) AS member_name, 
                CASE when m.recommendedby <> 0  then CONCAT_WS(", ", r.surname, r.firstname) 
                    when m.recommendedby = 0  then null end AS recommender_name
        FROM Members as m 
        INNER JOIN Members as r 
        ON m.recommendedby = r.memid
        Order By member_name
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks for Q10")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

__/* Q12: Find the facilities with their usage by member, but not guests */__


```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT f.facid, 
            f.name AS facility_name,
            b.memid,
            COUNT(b.memid) AS member_usage
        FROM Bookings b 
        LEFT JOIN Facilities f
        USING(facid)
        LEFT JOIN Members m 
        USING(memid)
        Group by b.memid, f.name 
        HAVING b.memid <> 0
        ORDER BY f.facid, m.memid
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks for ")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

__/* Q13: Find the facilities usage by month, but not guests */__

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT b.facid, 
            f.name AS facility_name,
            EXTRACT(MONTH FROM b.starttime) AS month,
            COUNT(EXTRACT(MONTH FROM b.starttime)) AS mo_usage
        FROM (SELECT *
              FROM Bookings 
              WHERE memid <> 0) AS b
        LEFT JOIN Facilities f
        ON b.facid = f.facid
        Group BY b.facid, EXTRACT(MONTH FROM starttime)
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
