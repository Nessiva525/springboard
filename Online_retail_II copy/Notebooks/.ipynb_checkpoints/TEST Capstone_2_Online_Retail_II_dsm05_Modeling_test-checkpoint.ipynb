**The Data**<br>
The Data is derived from the [Online Retail II](https://archive.ics.uci.edu/dataset/502/online+retail+ii) dataset from the UCI machine learning repository.  The original dataset was a collection of all online transations from a UK based retail company from 01/12/2009 to 09/12/2011. 
<br><br>
The data was was cleaned to remove all cancelled orders, any duplicates and other miscelaneous items that needed cleaning.  The data was then split into two dataframes, one with known customers and one with unknown customers.  The unknown customer data was set aside and the known customer data was grouped by customer and transformed into a Recency, Frequency, Monetary (RFM) dataset.   
<br>
In addition to basic RFM data, the rfm data was expanded by calculating Average Order Value, Customer Lifetime, and Customer Lifetime Value.  This expanded rfm dataset will be used for comparison of models.
<br><br>
**The Goal** <br>
The goal with this project is to find the distinct customer segments to to allow the company to better focus on marketing strategies for each segment.

## Model 1 - Basic RFM data
Model one will be rfm values only.

**Finding K for k-means**<br>
I will use the elbow method and the silhouette score to find the optimal k for the KMeans clustering.

```python
# create empty list for sum of squares values
inertia = []
k_values = range(1,10)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=25)
    kmeans.fit(rfm_scaled)
    
    # inertia_ for elbow method
    inertia.append(kmeans.inertia_)
    if k==1:
        print(f'k={k}, inertia: {inertia[-1]:.4f}')
    else:
        print(f"k={k}, inertia: {inertia[-1]:.4f}, change: {(inertia[-1]-inertia[0])/inertia[0]}")
    # print(inertia)
    
# plot elbow plot
plt.title('Model 1 Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(k_values, inertia, marker='o')    
plt.show()
```

Based on the elbow method, k=2 would likely be optimal but k=3 or k=4 could also be used .  Next I will check the silhouette_score to better decide which k to use.

```python
k_values = range(2,10)
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=25)
    labels = kmeans.fit_predict(rfm_scaled)
    score = silhouette_score(rfm_scaled, labels)
    silhouette_scores.append(score)
    print(f'k={k}, Silhouette Score: {score:.4f}')
    # print(labels)
```

The silhouette_scores suggests k=2 is the best seperation of the clusters.  This possibly could be a seperation between wholseale and non wholesale customers.  However, combined with the elbow method and the business goal of customer segmentation a k=2 might be too broad for customer segmentation purposes.  A k=4 has the next highest silhouette_score and is indicated by th elbow method as a possible optimal k measure.

Without guidance from managers in the company, I would pick a k=4 for this data based on the previous results. 

That being said, I will do a PCA to show the clusterings based on k=2 through k=5 below to visualize.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
rfm_scaled = pd.read_csv('../Data/rfm_scaled_df.csv', dtype='float32', index_col='Customer ID')


kmeans = KMeans(n_clusters=2, random_state=25)
kmeans.fit_predict(rfm_scaled)

# Assume 'rfm_scaled' is your normalized dataset and 'kmeans.labels_' contains the original K=2 clusters
rfm_scaled['Cluster'] = kmeans.labels_  # Assigning original K=2 cluster labels

# Separate Cluster 0 and Cluster 1
cluster_0 = rfm_scaled[rfm_scaled['Cluster'] == 0].drop(columns=['Cluster'])
cluster_1 = rfm_scaled[rfm_scaled['Cluster'] == 1].drop(columns=['Cluster'])

# Function to determine the optimal number of clusters within each segment
def find_optimal_k(data, max_k=6):
    inertia = []
    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, random_state=25, n_init=10)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    
    # Plot elbow curve
    plt.figure(figsize=(6,4))
    plt.plot(range(2, max_k + 1), inertia, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal K')
    plt.show()

# Determine best K for Cluster 0
print("Elbow Method for Cluster 0")
find_optimal_k(cluster_0)

# Determine best K for Cluster 1
print("Elbow Method for Cluster 1")
find_optimal_k(cluster_1)
```

The silhouette_score suggests that k=2 is optimal for the expanded dataset. The clusters in model 2 are less defined than the with model 1 even with the smaller number of clusters and the oprimal of k=2 provides less value for marketing. From the broad focus of customer segmentation for marketing, Model 1 using k=4 provides the best information.

While RFM is a good start, input from management in a business setting would allow for additional guidance as to how to improve the focus of the model and to optimize the features used to best meet the business's needs.

The number of customers in cluster 0 does not change between the two k numbers indicating it is a strog grouping of customers. There is no loss of value from for marketing because there is no change.  Cluster 1 from k=2 gets broken down further into different segments with k=4 which adds value for marketing.

The counts comparison adds additional backing to k=4 being the optimal cluster number.  The silhouette_score of 0.3820 for k=4  is lower than the silhouette_score of 0.4378 for k=2 but the added value for marketing could be worth it.
```python
## import with correct data types

# Set dtypes
dtype_dict={'Invoice':'string', 
            'StockCode':'string',
            'Description':'string',
            'Quantity':'int64',
            'Price':'float64',
            'Customer ID':'string',
            'Country':'category',
            'Sales':'float64'}
# import
has_id = pd.read_csv('../Data/has_id.csv', dtype = dtype_dict, parse_dates = ['InvoiceDate'])

# correct YearMonth to monthly period
has_id['YearMonth'] = pd.PeriodIndex(has_id['YearMonth'], freq='M')

has_id.info()
```

```python


# Ensure the data is sorted by customer and date
has_id = has_id.sort_values(by=["Customer ID", "InvoiceDate"])

# Define a function to compute rolling total sales
def rolling_sales(group, window):
    return group.rolling(window=window, on="InvoiceDate")["Sales"].sum()

# Compute rolling total sales for different time windows
has_id["Rolling_1M_Sales"] = has_id.groupby("Customer ID", group_keys=False).apply(rolling_sales, "30D")
has_id["Rolling_3M_Sales"] = has_id.groupby("Customer ID", group_keys=False).apply(rolling_sales, "90D")
has_id["Rolling_6M_Sales"] = has_id.groupby("Customer ID", group_keys=False).apply(rolling_sales, "180D")

# Display result
has_id.head()

```

```python
# Set reference date for frequency to one day after max date
max_date = has_id['InvoiceDate'].max()
reference_date = max_date + pd.Timedelta(days=1)
print(f"Max Date: {max_date}, Reference Date: {reference_date}")
```

```python
rfm2 = has_id.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  # Recency
    'Invoice': 'nunique',                                    # Frequency
    'Rolling_1M_Sales': 'mean',                                        # Monetary
    'Rolling_3M_Sales': 'mean',
    'Rolling_6M_Sales': 'mean',
    'Sales': 'sum' 
}).reset_index()
rfm2.rename(columns={
    'InvoiceDate': 'Recency',
    'Invoice': 'Frequency',
    'Rolling_1M_Sales': 'Monetary_1M',
    'Rolling_3M_Sales': 'Monetary_3M',
    'Rolling_6M_Sales': 'Monetary_6M',
    'Sales': 'CLV'
}, inplace=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.preprocessing import StandardScaler
```

```python
path = '../Data/rfm.csv'
dtype_dict= {'Recency': 'int32',
               'Frequency': 'int32',
               'Monetary': 'float32'}
rfm = pd.read_csv(path, dtype= dtype_dict, index_col='Customer ID')
```

```python
rfm.info()
```

```python
rfm.head()
```

```python


# rfm.columns #= ["Customer ID", "Recency", "Frequency", "Monetary"]

# Log transformation to reduce skewness
rfm= np.log1p(rfm)

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[["Recency", "Frequency", "Monetary"]])

# Perform Hierarchical Clustering
Z = linkage(rfm_scaled, method="ward")  # Ward's method minimizes variance

# Plot Dendrogram with the new cutoff at 50
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode="lastp", p=20, leaf_rotation=45, leaf_font_size=10)
plt.axhline(y=50, color='r', linestyle='--', label="Cutoff at 50")  # New cutoff at 50
plt.legend()
plt.title("Customer Segmentation - Dendrogram")
plt.xlabel("Customer ID")
plt.ylabel("Distance")
plt.show()

# Apply the new optimal cutoff distance
optimal_cutoff = 50
rfm["Cluster"] = fcluster(Z, optimal_cutoff, criterion="distance")  # Uses distance instead of maxclust

# Cluster Profiling
cluster_summary = rfm.groupby("Cluster").mean()
print("Cluster Summary:\n", cluster_summary)

# Visualizing clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=rfm, x="Recency", y="Monetary", hue="Cluster", palette="viridis")
plt.title("Customer Segments (Recency vs. Monetary)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Apply clustering using distance-based cutoff at 50
optimal_cutoff = 50
rfm["Cluster"] = fcluster(Z, optimal_cutoff, criterion="distance")

# Compute Silhouette Score
sil_score = silhouette_score(rfm_scaled, rfm["Cluster"])
print(f"Silhouette Score for Hierarchical Clustering (Cutoff at 50): {sil_score:.4f}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Compute Silhouette Score for Hierarchical Clustering
optimal_cutoff = 50
rfm["Cluster_HC"] = fcluster(Z, optimal_cutoff, criterion="distance")
sil_score_hc = silhouette_score(rfm_scaled, rfm["Cluster_HC"])
print(f"Silhouette Score for Hierarchical Clustering (Cutoff at 50): {sil_score_hc:.4f}")

# Step 2: Find optimal K for K-Means using Elbow Method
wcss = []
K_range = range(2, 11)  # Testing K from 2 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)  # Sum of squared distances

# Plot Elbow Method
plt.figure(figsize=(8, 4))
plt.plot(K_range, wcss, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal K in K-Means")
plt.show()

# Step 3: Apply K-Means with the optimal K (Assume K=4 from the Elbow Method)
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
rfm["Cluster_KMeans"] = kmeans.fit_predict(rfm_scaled)

# Step 4: Compute Silhouette Score for K-Means
sil_score_kmeans = silhouette_score(rfm_scaled, rfm["Cluster_KMeans"])
print(f"Silhouette Score for K-Means (K={optimal_k}): {sil_score_kmeans:.4f}")

# Step 5: Compare and Interpret Results
if sil_score_kmeans > sil_score_hc:
    print("K-Means performs better than Hierarchical Clustering based on Silhouette Score.")
else:
    print("Hierarchical Clustering performs better than K-Means based on Silhouette Score.")
```
