```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.capfriendly.com/browse/active/2024?stats-season=2024&display=signing-team,country,weight,height,signing-status,expiry-year,caphit-percent,aav,length,skater-individual-advanced-stats,skater-on-ice-advanced-stats,goalie-advanced-stats,signing-age,signing-date,extension&pg=1"
```

```python
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html')
```

```python
# Find table
table = soup.find('table')
```

```python
table
```

```python
#find table titles

world_titles = soup.find_all('th')
```

```python
world_titles
```

```python
#extract table titles

world_table_titles = [title.text for title in world_titles]
```

```python
world_table_titles
```

```python
#add table titles to df

df = pd.DataFrame(columns = world_table_titles)

df
```

```python
#collect row data

column_data = soup.find_all('tr')
```

```python
#extract table data and add to df

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
#All 32 pages of active player data from Capfriendly.com for the 2023-24 season

page = 1

url_test = f"https://www.capfriendly.com/browse/active/2024?stats-season=2024&display=signing-team,country,weight,height,signing-status,expiry-year,caphit-percent,aav,length,skater-individual-advanced-stats,skater-on-ice-advanced-stats,goalie-advanced-stats,signing-age,signing-date,extension&pg={page}"
r = requests.get(url_test)
soup = BeautifulSoup(r.text, 'html')
table = soup.find('table')
world_titles = soup.find_all('th')
world_table_titles = [title.text for title in world_titles]
all_player_data = pd.DataFrame(columns = world_table_titles)

while page!= 33:
    url_test = f"https://www.capfriendly.com/browse/active/2024?stats-season=2024&display=signing-team,country,weight,height,signing-status,expiry-year,caphit-percent,aav,length,skater-individual-advanced-stats,skater-on-ice-advanced-stats,goalie-advanced-stats,signing-age,signing-date,extension&pg={page}"
    r = requests.get(url_test)
    soup = BeautifulSoup(r.text, 'html')
    #table = soup.find('table')
    #world_titles = soup.find_all('th')
    #world_table_titles = [title.text for title in world_titles]
    #df_test = pd.DataFrame(columns = world_table_titles)
    column_data = soup.find_all('tr')
    
    for row in column_data[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text for data in row_data]
    
        length = len(all_player_data)
        all_player_data.loc[length] = individual_row_data
    
    page = page+1
```
